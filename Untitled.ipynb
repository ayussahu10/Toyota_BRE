````python
def retreive_relevant_docs(chromdb_collection,prompt):
    title_query_result = chromdb_collection.query(
        query_texts=prompt,
        n_results=10
    )
    return title_query_result
````

````python
def get_response(prompt, system, temperature = 1.0):
    response = openai.ChatCompletion.create(
        model = "gpt-3.5-turbo",
        messages = [{"role": "user", "content": prompt},
                {"role": "system", "content": system}],
        temperature = temperature
    )
    return response['choices'][0]['message']['content']
````

````python
def create_context(df):
    context = ''
    for i,doc in enumerate(df['documents']):
        context += f"Document {i+1}:\n{doc}\n"
    return context
````

````python
user_query = 'Provide a list of cursor names used in the code.'
````

````python
title_query_result = retreive_relevant_docs(chromdb_collection,[user_query])

context = create_context(title_query_result)
# print(context)

# generate RAG responses by passing top relevant documents
system_prompt = """You are an intelligent question answer assistant. Your task is to answer user 
questions based on the context provided to you. Display the reason of your answers along with 
accuracy score as well."""

message_prompt = f"""
Question:
```
{user_query}
```

Context:
```
{context}
```

Answer:

"""

response = get_response(message_prompt, system_prompt, temperature = 0.9)
````
